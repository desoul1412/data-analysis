A **variable** is like a labeled box that stores data. Python automatically figures out the type.

```python
scores = [450, 320, 580, 290, 610, 400, 550]

count = 0
max_score = 0
for score in scores:
    if score >= 500:
        count += 1
        print(f"High score: {score}")
    else:
        print(f"Regular score: {score}")
    
    if score > max_score:
        max_score = score

print(f"\nTotal high scores (>=500): {count}")
print(f"Highest score: {max_score}")
```

```python
def calculate_player_stats(playtime_minutes,levels_completed,coins_earned):
    efficiency = coins_earned / playtime_minutes
    avg_level_time = playtime_minutes / levels_completed
    result = (efficiency,avg_level_time)
    if efficiency > 10 and avg_level_time < 5:
        return result + ("Excellent",)
    elif efficiency > 5:
        return result + ("Good",)
    else:
        return result + ("Needs Improvement",)

result = calculate_player_stats(60, 15, 750)
print(f"Efficiency: {result[0]} coins/min")
print(f"Avg Level Time: {result[1]} min/level")
print(f"Performance: {result[2]}")
```

```python
players = [
    {"name": "Alice", "level": 25, "playtime": 120, "premium": True},
    {"name": "Bob", "level": 15, "playtime": 60, "premium": False},
    {"name": "Charlie", "level": 30, "playtime": 180, "premium": True},
    {"name": "Diana", "level": 10, "playtime": 45, "premium": False},
    {"name": "Eve", "level": 28, "playtime": 150, "premium": True}
]

def efficiency_calc(level,playtime):
    return level/playtime

premium_players = []
highest_eff_player = {"name":"","efficiency":0}
player_levels = []
for player in players:
    efficiency = efficiency_calc(player["level"],player["playtime"])
    player["efficiency"] = efficiency

    if player["premium"] == True:
        premium_players.append(player["name"])
    
    for name, efficiency in player.items():
        if player["efficiency"] > highest_eff_player["efficiency"]:
            highest_eff_player["name"] = player["name"]
            highest_eff_player["efficiency"] = player["efficiency"]
    
    player_levels.append(player["level"])

print(f"Premium players: {premium_players}")
print(f"Most efficient player: {highest_eff_player["name"]} ({highest_eff_player["efficiency"]} level/min)")
print(f"Average level: {sum(player_levels)/len(player_levels)}")


```
