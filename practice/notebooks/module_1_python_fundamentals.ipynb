## 1.1 Variables and Data Types

### ðŸ“š Concept

A **variable** is like a labeled box that stores data. Python automatically figures out the type.

**Common types in game analytics:**
- `int` (integer): Whole numbers â†’ `level = 5`
- `float`: Decimals â†’ `playtime = 45.5`
- `str` (string): Text â†’ `player_id = "P001"`
- `bool` (boolean): True/False â†’ `is_premium = True`

```python
# Example: Player metrics
player_id = "P001"
level_reached = 7
playtime_hours = 2.5
is_paying_user = True

# Check the type
print(type(player_id))
print(type(level_reached))
print(type(playtime_hours))
print(type(is_paying_user))
```
